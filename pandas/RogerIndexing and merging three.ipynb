```python
# because the lengths are the same I should be able to apply one index to another
# the question to answer is what, when, where and who
# so the date index is appropriate
# take the index from d and appy to e
e_i = d.index
e.set_index(e_i, inplace=True)
e.iloc[:5]
```

```python
# okay so this should work
# the indexs are the same so now stick e onto d along axis one
d_e = pd.concat([d,e], axis=1)
d_e.iloc[:5]
```

```python
# okay!
# wirte a function that attributes cost to the persons choice of restaurant
# by taking a list of names and outputting directly to the df
# becasue now the df is combined

def get_cost(df_cost, x, y):
    d ={}
    # x is a list of names:
    for i, n in enumerate(x):
        a = []
        b = list(df_cost[n])
        name = n + '_cost'
        #print(b)
        for j, o in enumerate(b):
            c = df_cost.iloc[j, y[o]]
            a.append(c)
        d.update({name:a})
    for key, value in d.items():
        
        d_e[key] = value
    return(d)
try_this = get_cost(d_e, ['Sheila', 'Dave', 'Pat'], rest)
```

```python
# do this with a dataframe:
values = np.arange(12).reshape((4,3))
index= [['a', 'a', 'b', 'b'], [1, 2, 1, 2]]
columns = [['Ohio', 'Ohio', 'Colorado'], ['Green', 'Red', 'Green']]
g = pd.DataFrame(values, index=index, columns=columns)
```

```python
# give the indexes some names
g.index.names = ['key1', 'key2']
g.columns.names = ['state', 'color']
g
```

```python
# so we can index by columns in this sense:
g['Ohio']
```

```python
# or swap levels if need be:
g.swaplevel('key1', 'key2')
```

```python
# or sort on an index:
print(g.sort_index(level=1))
# remember it looks like this
print('the original df:')
print(g)
```

```python
# you can combine swap and sort:
g.swaplevel(0,1).sort_index(level=0)
```

```python
g_1 = pd.DataFrame({'key':['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
g_2 = pd.DataFrame({'key':['a', 'b', 'd'], 'data2':range(3)})
pd.merge(g_1, g_2)
```

```python
g_1 = pd.DataFrame({'lkey':['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
g_2 = pd.DataFrame({'rkey':['a', 'b', 'd'], 'data2':range(3)})
pd.merge(g_1, g_2, left_on='lkey', right_on='rkey')
```

```python
# looking at the result of the merge above
# and the orginal dfs there is missing data
# and added data.....
print(g_1)
print(g_2)
```
